```python
def print_stat(stats_object, param_name):
    print(param_name)
    print(stats_object.parWithName(param_name).limits[0])
    print(stats_object.parWithName(param_name).limits[1])
    print(stats_object.parWithName(param_name).mean)
```

```python
def print_stats(stats_object):
    print_stat(stats_object,'logGnu')
    print_stat(stats_object,'logA')
    print_stat(stats_object,'ns')
    print_stat(stats_object,'theta_MC_100')
    print_stat(stats_object,'ombh2')
    print_stat(stats_object,'omch2')
    print_stat(stats_object,'tau')
    print_stat(stats_object, 'H0')
    print_stat(stats_object, 'As')
```

```python
#Import set of single samples
import numpy as np
import math
chains = []
for i in range(10):
    data = np.loadtxt('/Users/katherinefraser/Documents/Physics/Cosmology_[P212]/Final_Project/Final_Code/cobaya_examples/chains/neutrinos.{}.txt'.format(i+1),skiprows=1)
    chains.append(data)
    print("loaded", i+1)
```

```python
#Set additional burn_in
burn_in = 0.1
total_samples = 0
for i in range(10):
    print(i, ", ", chains[i].shape[0])
    total_samples += chains[i].shape[0]
    burned_samples = math.floor(chains[i].shape[0]*burn_in)
    chains[i] = chains[i][burned_samples:, :]
print("total samples: ", total_samples)
```

```python
#Define Gelman-Rubin criterion for given parameter for each chain (split chain into four parts and then compare)
def GR(param_array):
    split = np.array_split(param_array,4)
    mean = [0, 0, 0, 0]
    var = [0, 0, 0, 0]
    for i in range(4):
        mean[i] = split[i].mean()
        var[i] = split[i].var()  
    GR = (np.asarray(mean).var())/(np.asarray(var).mean())    
    return GR
```

```python
#Compute GR criterion for the individual parameters for the combined set of samples
all_chains = np.concatenate((chains[:]))
logGnu = GR(all_chains[:,2])
logA = GR(all_chains[:,3])
ns = GR(all_chains[:,4])
thetamc = GR(all_chains[:,5])
ombh2 = GR(all_chains[:,6])
omch2 = GR(all_chains[:,7])
tau = GR(all_chains[:,8])
Aplanck = GR(all_chains[:,9])
print("logGnu: ", logGnu)
print("logA: ", logA)
print("ns: ", ns)
print("thetamc: ", thetamc)
print("ombh2: ", ombh2)
print("omch2: ", omch2)
print("tau: ", tau)
print("Aplanck: ", Aplanck)
```

```python
#Compute GR values for different parameters
for i in range(10):
    print("chain number: ", i+1)
    #Compute GR for each parameter
    logGnu = chains[i][:,2]
    GRvalue = GR(logGnu)
    print("    logGnu: ", GRvalue)
    logA = chains[i][:,3]
    GRvalue = GR(logA)
    print("    logA: ", GRvalue)
    ns = chains[i][:,4]
    GRvalue = GR(ns)
    print("    ns: ", GRvalue)
    thetaMC = chains[i][:,5]
    GRvalue = GR(thetaMC)
    print("    thetaMC: ", GRvalue)
    ombh2 = chains[i][:,6]
    GRvalue = GR(ombh2)
    print("    ombh2: ", GRvalue)
    omch2 = chains[i][:,7]
    GRvalue = GR(omch2)
    print("    omch2: ", GRvalue)
    tau = chains[i][:,8]
    GRvalue = GR(tau)
    print("    tau: ", GRvalue)
```
